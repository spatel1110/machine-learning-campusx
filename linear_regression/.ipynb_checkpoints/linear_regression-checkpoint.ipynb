```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('./../data/IPL Ball-by-Ball 2008-2020.csv')
```

```python
df
```

```python
runs = df[df['batsman'] == 'V Kohli']
```

```python
total_run = runs.groupby('id').sum('batsman_runs')
total_ball = runs.groupby('id').count()
```

```python
plt.scatter(total_ball['ball'],total_run['batsman_runs'])
plt.xlabel('Total ball in each match')
plt.ylabel('Total runs in each match')
plt.show()
```

```python
X = total_ball.iloc[:,2:3]
Y = total_ball.iloc[:,6:7]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, Y, test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train.values,y_train.values)
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
lr.predict(X_train.values)
```

```python
plt.scatter(total_ball['ball'],total_run['batsman_runs'])
plt.plot(X_train,lr.predict(X_train.values),color='red')
plt.xlabel('Ball')
plt.ylabel('Runs')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = lr.predict(X_test.values)
```

```python
X_test.values
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
print("MSE",r2_score(y_test,y_pred))
r2 = r2_score(y_test,y_pred)
```
